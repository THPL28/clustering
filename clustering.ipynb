# 🚀 Clustering com K-Means

Este projeto demonstra, na prática, como aplicar **clustering (aprendizado não supervisionado)** utilizando o algoritmo K-Means.

O objetivo é entender como os dados podem ser agrupados com base em similaridade, sem o uso de rótulos.

---

## 📦 1. Importação de bibliotecas

    import numpy as np
    import matplotlib.pyplot as plt
    from sklearn.cluster import KMeans
    from sklearn.datasets import make_blobs

### 🔍 Explicação

- numpy: manipulação de arrays e operações matemáticas  
- matplotlib: visualização dos dados  
- KMeans: algoritmo de clustering  
- make_blobs: gera dados sintéticos para simulação  

---

## 🎲 2. Geração de dados sintéticos

    X, y = make_blobs(n_samples=300, centers=4, cluster_std=1.2, random_state=42)

### 🔍 Explicação

- n_samples=300: número total de pontos  
- centers=4: quantidade de clusters (grupos)  
- cluster_std=1.2: controla a dispersão dos pontos  
- random_state=42: garante reprodutibilidade  

📌 Simula dados reais (clientes, produtos, etc.)

---

## 🤖 3. Aplicando o K-Means

    kmeans = KMeans(n_clusters=4, random_state=42)
    kmeans.fit(X)

### 🔍 Explicação

- n_clusters=4: número de grupos  
- fit(): treina o modelo  

📌 Funcionamento:
1. Inicializa centróides  
2. Atribui pontos ao mais próximo  
3. Recalcula centróides  
4. Repete até convergir  

---

## 🏷️ 4. Resultados

    labels = kmeans.labels_
    centroids = kmeans.cluster_centers_

### 🔍 Explicação

- labels: cluster de cada ponto  
- centroids: centro dos clusters  

---

## 📊 5. Visualização

    plt.scatter(X[:, 0], X[:, 1], c=labels)
    plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200)
    plt.title("Clustering com K-Means")
    plt.show()

### 🔍 Explicação

- Cada cor = um cluster  
- "X" = centróides  

---

## 📉 6. Elbow Method

    inertia = []
    for k in range(1, 10):
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(X)
        inertia.append(kmeans.inertia_)

    plt.plot(range(1, 10), inertia)
    plt.title("Elbow Method")
    plt.show()

### 🔍 Explicação

- inertia_: soma das distâncias aos centróides  
- Objetivo: encontrar o "cotovelo"  

---

## 🧠 Conclusão

- K-Means é simples e eficiente  
- Funciona melhor com clusters bem definidos  
- Escolher K corretamente é essencial

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=300, centers=4, cluster_std=1.2, random_state=42)
kmeans = KMeans(n_clusters=4, random_state=42, n_init='auto')
kmeans.fit(X)

labels = kmeans.labels_
centroids = kmeans.cluster_centers_

plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200)
plt.title("Clustering com K-Means")
plt.show()
```

```python
inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 10), inertia)
plt.title("Elbow Method")
plt.show()
```
